```python
class Graph:
    def __init__(self, size=0):
        self.size = size
        self.items = [[0 for _ in range(size)] for _ in range(size)]

    def is_empty(self):
        return self.size == 0

    def insert_vertex(self):
        for i in range(self.size):
            self.items[i].append(0)
        self.items.append([0 for _ in range(self.size + 1)])
        self.size += 1
        print("Vertex added.")

    def insert_edge(self, vi, vj):
        if vi >= self.size or vj >= self.size:
            print("Invalid vertex index")
            return 
        self.items[vi][vj] = 1
        self.items[vj][vi] = 1
        print(f"Edge added between v{vi} and v{vj}.")

    def delete_vertex(self, v):
        if v >= self.size:
            print("Vertex not found")
            return
        for i in range(self.size):
            del self.items[i][v]
        del self.items[v]
        self.size -= 1
        print("Vertex removed.")

    def delete_edge(self, vi, vj):
        if vi >= self.size or vj >= self.size:
            print("Vertex not found")
            return
        self.items[vi][vj] = 0
        self.items[vj][vi] = 0
        print("Edge removed.")

    def display(self):
        print("Adjacency Matrix:")
        for i in range(self.size):
            print("v{} |".format(i), end=" ")
            for j in range(self.size):
                print(self.items[i][j], end=" ")
            print("|")


# Menu-driven interface
g = Graph()

while True:
    print("\n==== Graph Operations ====")
    print("1. Insert Vertex")
    print("2. Insert Edge")
    print("3. Delete Vertex")
    print("4. Delete Edge")
    print("5. Display Graph")
    print("6. Check if Graph is Empty")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == '1':
        g.insert_vertex()

    elif choice == '2':
        vi = int(input("Enter index of first vertex: "))
        vj = int(input("Enter index of second vertex: "))
        g.insert_edge(vi, vj)

    elif choice == '3':
        v = int(input("Enter index of vertex to delete: "))
        g.delete_vertex(v)

    elif choice == '4':
        vi = int(input("Enter index of first vertex: "))
        vj = int(input("Enter index of second vertex: "))
        g.delete_edge(vi, vj)

    elif choice == '5':
        g.display()

    elif choice == '6':
        if g.is_empty():
            print("Graph is empty.")
        else:
            print("Graph is not empty.")

    elif choice == '7':
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
class Graph:
    def __init__(self, vno):
        self.vertices = vno
        self.adj_matrics = [[0] * vno for _ in range(vno)]

    def add_edge(self, u,v,weight=1):
        if 0<=u<self.vertices and 0<=v<self.vertices:
            self.adj_matrics[u][v] = weight
            self.adj_matrics[v][u] = weight
        else:
            print("Invalid")

    def remove_edge(self, u,v,weight=1):
        if 0<=u<self.vertices and 0<=v<self.vertices:
            self.adj_matrics[u][v] = 0
            self.adj_matrics[v][u] = 0
        else:
            print("Invalid")

    def has_edge(self, u,v,):
        if 0<=u<self.vertices and 0<=v<self.vertices:
            return self.adj_matrics[u][v] != 0
        else:
            print("Invalid")


    def print_adj_matrix(self):
        for row in self.adj_matrics:
            print(" ".join(map(str,row)))

g = Graph(5)

g.add_edge(0,1)
g.add_edge(1,2)
g.add_edge(1,3)
g.add_edge(2,3)
g.add_edge(3,4)

g.print_adj_matrix()

g.remove_edge(1,2)
print("\n")
g.print_adj_matrix()
```
