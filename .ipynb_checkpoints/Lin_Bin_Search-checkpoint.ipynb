```python
### Detailed Notes on Linear and Binary Search with Steps Count

#### Data Preparation
'''python
nums = [i for i in range(1000)]  # Creating a list of integers from 0 to 999
target = 900  # Target value to search
'''
- **`nums`**: A list containing numbers from 0 to 999.
- **`target`**: The value we are trying to locate within the list.

---

#### Linear Search Function
'''python
def linearSearch1(nums, target):
    steps = 0
    for i in range(0, len(nums)):
        steps += 1
        if nums[i] == target:
            print(f"Steps taken for Linear Search = {steps}")
            return i
    else:
        print(f"Steps taken for Linear Search = {steps}")
        return -1
'''
- **Purpose**: To find the `target` in the list using a simple sequential search.
- **Logic**:
  - Loop through the list from start to end.
  - Increment `steps` with each iteration to track the number of comparisons made.
  - If the current element matches the `target`, return the index and print the steps taken.
  - If the loop completes without finding the `target`, print the steps and return `-1`.

---

#### Binary Search Function
'''python
def binarySearch1(nums, target, left, right):
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        
        if nums[mid] == target:
            print(f"Steps taken for Binary Search = {steps}")
            return mid
        elif nums[mid] < target:
            left = mid + 1 
        else:
            right = mid - 1 
    else:
        print(f"Steps taken for Binary Search = {steps}")
        return -1
'''
- **Purpose**: To find the `target` in the list using the binary search technique.
- **Logic**:
  - Divide the list into halves by calculating the middle index (`mid = (left + right) // 2`).
  - Compare the middle element with the `target`:
    - If it matches, return the index and print the steps.
    - If the middle element is less than the `target`, search the right half (`left = mid + 1`).
    - If it is greater, search the left half (`right = mid - 1`).
  - Repeat the process until the `target` is found or the range becomes invalid.
  - Print the steps and return `-1` if the `target` is not found.

---

#### Perform Searches
'''python
left = 0
right = len(nums) - 1

result1 = linearSearch1(nums, target)  # Linear Search
result2 = binarySearch1(nums, target, left, right)  # Binary Search
'''
- **left**: The starting index for binary search.
- **right**: The ending index for binary search.
- Perform both linear and binary searches for the `target` and store their results.

---

#### Output Results
'''python
if result1 != -1:
    print(f"Value Found at index {result1} using Linear Search")
else:
    print("Value not found using Linear Search")

if result2 != -1:
    print(f"Value Found at index {result2} using Binary Search")
else:
    print("Value not found using Binary Search")
'''
    
- **Purpose**: To display the results of both search methods.
- **Logic**:
  - If the result of linear or binary search is not `-1`, print the index where the `target` was found.
  - If the result is `-1`, print that the value was not found.

---

### Key Differences Between Linear and Binary Search
| Aspect            | Linear Search                  | Binary Search                  |
|-------------------|-------------------------------|--------------------------------|
| **Approach**      | Sequentially checks elements. | Divides the list into halves.  |
| **Efficiency**    | O(n)                          | O(log n)                       |
| **Precondition**  | No specific requirements.     | Requires a sorted list.        |
| **Steps Taken**   | Depends on the target position | Logarithmic steps in sorted data |

---

### Example Execution
- **Target**: `900`
- **Linear Search Steps**: Prints the number of steps taken to find the value sequentially.
- **Binary Search Steps**: Prints the number of steps taken to find the value using divide-and-conquer.

### Expected Output
'''text
Steps taken for Linear Search = 901
Value Found at index 900 using Linear Search

Steps taken for Binary Search = 10
Value Found at index 900 using Binary Search
'''
```

```python
nums = [i for i in range(1000)]
target = 900

def linearSearch1(nums, target):
    steps = 0
    for i in range(0,len(nums) - 1):
        steps += 1
        if nums[i] == target:
            print(f"Steps taken of Linear = {steps}")
            return i
    else:
        print(f"Steps taken of Linear = {steps}")
        return -1


left = 0
right = len(nums) - 1
def binarySearch1(nums, target, left, right):
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        
        if nums[mid] == target:
            print(f"Steps taken of Binary = {steps}")
            return mid
        elif nums[mid] < target:
            left = mid + 1 
        else:
            right = mid - 1 
    else:
        print(f"Steps taken of Binary = {steps}")
        return -1


    
result1 = linearSearch1(nums, target)
result2 = binarySearch1(nums,target, left, right)
if result1 != -1:
    print(f"Value Found at index {result1}")
else:
    print("Value not found")
```
